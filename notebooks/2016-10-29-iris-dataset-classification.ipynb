```python
"""
2016-10-26 9:49:00
Deep Learning Book p. 98-106

LOG: Today I installed a "data science" toolkit, including scikit-learn, jupyter notebook, and theano after reading
part of the introduction on "Machine Learning Basics".

LEARNING: A supervised learning tutorial using the Iris dataset
"""
from sklearn import datasets
iris = datasets.load_iris()
```

```python
from sklearn.svm import SVC
clf = SVC()
clf.fit(iris.data, iris.target_names[iris.target])
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
sepal_length = iris.data[:, 0]
sepal_width = iris.data[:, 1]
predicted = list(clf.predict(iris.data))
actual = iris.target_names[iris.target]
right_or_wrong = (predicted[i] == actual[i] for i in range(len(predicted)))
colors = ['lightgreen' if right else 'red' for right in right_or_wrong]

plt.figure(2, figsize=(8,6))
plt.clf()

# Plot the training points and whether we got the predictions correct
plt.scatter(sepal_length, sepal_width, c=colors)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(sepal_length.min() - .5, sepal_length.max() + .5)
plt.ylim(sepal_width.min() - .5, sepal_width.max() + .5)
plt.xticks(())
plt.yticks(())

plt.show()
```
